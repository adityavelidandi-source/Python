```python
num=input("enter number to check")
num=int(num)
if(num%2==0):
  print(f"number {num} is even")
else:
  print(f"number {num} is odd")
```

Let's break down the code step by step:

```python
# This line prompts the user to "enter number to check" and stores their input as a string in the variable `num`.
num=input("enter number to check")

# This line converts the string input received from the user into an integer. This is necessary for performing mathematical operations.
num=int(num)

# This is an `if` statement that checks if the number is even.
# The modulo operator (`%`) returns the remainder of a division.
# If `num % 2` is `0`, it means `num` is perfectly divisible by 2, and therefore even.
if(num%2==0):
  # If the condition in the `if` statement is true (the number is even), this line is executed.
  # It prints a formatted string indicating that the number is even.
  print(f"number {num} is even")
# The `else` statement is executed if the condition in the `if` statement is false (the number is not even, meaning it's odd).
else:
  # If the number is odd, this line prints a formatted string indicating that the number is odd.
  print(f"number {num} is odd")
```

write python code for calculator application. No functions. Two inputs will be given as inputs by user. no if-else conditions. print all arthematic operation values.

```python
print("--- Calculator Application ---")

# Get the first number from the user
num1_str = input("Enter the first number: ")
num1 = float(num1_str) # Convert input string to a float for arithmetic operations

# Get the second number from the user
num2_str = input("Enter the second number: ")
num2 = float(num2_str) # Convert input string to a float

print("------------------------------")
print("Results of Arithmetic Operations:")
print("------------------------------")

# Addition
print(f"{num1} + {num2} = {num1 + num2}")

# Subtraction
print(f"{num1} - {num2} = {num1 - num2}")

# Multiplication
print(f"{num1} * {num2} = {num1 * num2}")

# Division
print(f"{num1} / {num2} = {num1 / num2}")

# Modulo (remainder)
print(f"{num1} % {num2} = {num1 % num2}")

# Exponentiation
print(f"{num1} ** {num2} = {num1 ** num2}")

# Floor Division (integer division)
print(f"{num1} // {num2} = {num1 // num2}")
```
